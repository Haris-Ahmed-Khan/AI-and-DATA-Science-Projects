```python
lab_columns = ['HB', 'TLC', 'PLATELETS', 'GLUCOSE', 'UREA', 'CREATININE']

for col in lab_columns:
    df[col] = df[col].fillna(df[col].median())


df.drop(columns=['BNP', 'EF'], inplace=True)

df.isnull().sum().sum()  # should be 0
```

```python
lab_columns = ['HB', 'TLC', 'PLATELETS', 'GLUCOSE', 'UREA', 'CREATININE']

for col in lab_columns:
    df[col] = pd.to_numeric(df[col], errors='coerce')

for col in lab_columns:
    df[col] = df[col].fillna(df[col].median())

df.drop(columns=['BNP', 'EF'], inplace=True)
```

```python
total_admissions=len(df)
print("Total Admissions :",total_admissions)
```

```python
bins=[0,18,35,50,65,80,100]
labels=['0-18','19-35','36-50','51-65','66-80','81+']
df["age_group"]=pd.cut(df["AGE"],bins=bins,labels=labels)
df['age_group'].value_counts().sort_index().plot(kind='bar')
plt.title("Admissions by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Count")
plt.show()
```

```python
df['TYPE OF ADMISSION-EMERGENCY/OPD'].value_counts().plot(kind='bar')
plt.title("Admission Type Distribution")
plt.show()
```

```python
df.groupby('OUTCOME')['DURATION OF STAY'].mean().plot(kind='bar')
plt.title("Avg Stay Duration by Outcome")
plt.ylabel("Days")
plt.show()
```

```python
gender_counts=df["GENDER"].value_counts()
print(gender_counts)
gender_counts.plot(kind='pie', autopct='%1.1f%%', title='Gender Distribution')
plt.ylabel('')
plt.show()
```
